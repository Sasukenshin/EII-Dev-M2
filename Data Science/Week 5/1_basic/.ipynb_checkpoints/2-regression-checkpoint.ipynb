## Part 2. Regression on Tabular data

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Here is an example with dummy data, 3 classes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
X = np.arange(-3,8,0.01)
```

```python
y = X**3 -9*X**2 + 5* X -1 + np.random.normal(0,10, len(X))
```

```python
plt.scatter(X,y, c=y)
plt.show()
```

**Question 1. Split data into train/test with ratio 2:1 ussing Sklearn train_test_split**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/4)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

Here is our model with 3 layers

### Architecture

```python
model = Sequential()
model.add(Dense(6, input_dim=1, activation='tanh')) ## Activations: 'relu', 'tanh', 'sigmoid'
model.add(Dense(4, activation='tanh'))
# model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='linear'))

print(model.summary())
```

### Update rule

```python
model.compile(loss='mae',
              optimizer='adam',
              metrics=['mae'])
```

```python
model.fit(X_train, y_train,
          epochs=1000,
          batch_size=64)
```

```python
_, train_err = model.evaluate(X_train, y_train, batch_size=64)
_, test_err = model.evaluate(X_test, y_test, batch_size=64)
print('train_err', train_err)
print('test_err', test_err)
```

Get prediction on train data

```python
y_train_pred = model.predict(X_train, batch_size=64)
plt.scatter(X_train, y_train)
plt.scatter(X_train,y_train_pred)
plt.show()
```

Get prediction on test data

```python
y_test_pred = model.predict(X_test, batch_size=64)
plt.scatter(X_test,y_test)
plt.scatter(X_test,y_test_pred)
plt.show()
```
