<h3>Import necessary modules</h3>

```python
import requests
from bs4 import BeautifulSoup
import time
```

## The http request

```python
url = "https://pokemondb.net/pokedex/all"
response = requests.get(url)
if response.status_code == 200:
    print("Success")
else:
    print("Failure")
```

```python
results_page = BeautifulSoup(response.content,'lxml')
print(results_page.prettify())
```

### Q1. Mine the table in the main page and save as a dataframe

```python
import pandas as pd
```

Get pokedex body

```python
pokedex = results_page.find("table", {"id": "pokedex"})
body = pokedex.find("tbody")
body
```

```python
cols = pokedex.find_all("th")
col_list = [col.get_text() for col in cols]
col_list
```

```python
rows = body.find_all("tr")
print(rows)
```

```python
print(len(rows)) # There are 890 pokemons
```

```python
list(rows[0].get_text().split('\n'))
```

```python
rows[0]
```

```python
cleaned_rows = []
for row in rows:
    cells = row.find_all("td")
    row_texts = [cell.get_text() for cell in cells]
    cleaned_rows.append(row_texts)
print(cleaned_rows)
```

```python
pokemon_df = pd.DataFrame(cleaned_rows)
pokemon_df.columns = col_list
pokemon_df
```

### Q2. Go to the page of each pokemon, and mine all further interesting information and add to the dataframe

Note: intersting information that can be render in a df cell aren't so many, so we focus on abilities

```python
def get_details(res_page):
    # Abilities
    vitals_table = res_page.find('table', {'class': 'vitals-table'})
    vitals_rows = vitals_table.find('tbody').find_all('tr')
    abilities = vitals_rows[5].find('span').get_text()
    return {"Abilities": abilities}
```

```python
pokemon_df["Abilities"] = ""

base_url = "https://pokemondb.net"

for index, row in enumerate(rows):
    url = base_url + row.find("a").get("href")
    time.sleep(5)
    response = requests.get(url)
    res_page = BeautifulSoup(response.content, 'lxml')
    # Get datas from details page
    details = get_details(res_page)
    #pokemon_df.set_value(index, 'Abilities', details['Abilities'])  # Deprecated
    pokemon_df.iat[index, 'Abilities'] = details['Abilities']
```
