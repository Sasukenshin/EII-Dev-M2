### Step 4. Set PassengerId as the index 

```python
df = pd.DataFrame(titanic)
df.set_index('PassengerId', inplace=True)
df
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
# sum the instances of males and females
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

# put them into a list called proportions
proportions = [males, females]

# Create a pie chart
plt.pie(
    # using proportions
    proportions,
    
    # with the labels being officer names
    labels = ['Males', 'Females'],
    
    # with no shadows
    shadow = False,
    
    # with colors
    colors = ['blue','red'],
    
    # with one slide exploded out
    explode = (0.15 , 0),
    
    # with the start angle at 90%
    startangle = 90,
    
    # with the percent listed as a fraction
    autopct = '%1.1f%%'
    )

# View the plot drop above
plt.axis('equal')

# Set labels
plt.title("Sex Proportion")

# View the plot
plt.tight_layout()
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
fare_by_age = titanic.groupby(['Fare', 'Age']).sum()
fare_by_age

s_plot = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex')
s_plot.set(title = 'Paid fare by age')

```

### Step 7. How many people survived?

### Step 8. Create a histogram with the Fare payed

```python
fare_df = titanic.Fare.sort_values(ascending = False)

# create the plot
plt.hist(fare_df)

# Set the title and labels
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Price Histogram')

# show the plot
plt.show()
```
