Q. Write linear regression function to predict medv using lstat and age

```r
attach(Boston)
```

```r
lm_medv = lm(medv~lstat + age, data = Boston) #Create the linear regression
summary(lm_medv) #Review the results
```

Q. Write linear regression function to predict medv using all others

```r
lm_medv = lm(medv~., data = Boston) #Create the linear regression
summary(lm_medv) #Review the results
```

Q. Write linear regression function to predict medv using all others except age

```r
lm_medv = lm(medv~. - lstat - age, data = Boston) #Create the linear regression
summary(lm_medv) #Review the results
```

Q. Write polynomial regression function predict medv using lstat and age

```r
lm_medv = lm(medv ~ polym(lstat, age, degree=2, raw=TRUE), data = Boston)
summary(lm_medv)
```
