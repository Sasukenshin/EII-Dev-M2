## Neural network

Q. Load Concrete dataset in data

```r
df <- read.csv("data/concrete.csv")
```

```r
head(df)
```

```r
attach(df)
```

```r
normalize <- function(x) {
  return ((x - min(x)) / (max(x) - min(x)))
}

df <- as.data.frame(lapply(df, normalize))
```

Q. Seperate data in train, test with ratio 2/1

```r
set.seed(1)
train_index <- sample(1:nrow(df), 0.7 * nrow(df))
test_index <- setdiff(1:nrow(df), train_index)
df_train = df[train_index,]
df_test = df[test_index,]
dim(df_train)
dim(df_test)
```

Q. Install and load package "neuralnet"

```r
install.packages('neuralnet')
library(neuralnet)
```

```r
?neuralnet
```

Q. Build a neural net to predict strength over all other data

```r
# nn <- neuralnet(strength ~ ., df_train, linear.output = TRUE, hidden = c(10,1))
nn <- neuralnet(strength ~ ., df_train)
nn
```

Q. Get the result on test set

```r
?compute
```

```r
comp <- compute(nn, df_test)
comp
```

```r
pred <- predict(nn, df_test)
table(df_test$strength, pred[, 1] > 0.5)
```

Q. Print the correlation between predicted strength and ground truth of test data

```r
cor(pred, df_test$strength)
```

```r
?predict
```

Q. Print the mean square error of your prediction

```r
comp$net.result
```

```r
results <- data.frame(actual = df_test$strength, prediction = comp$net.result)
results
```

```r
mse <- sum((results$actual - results$prediction)^2)/nrow(results)
mse
```
