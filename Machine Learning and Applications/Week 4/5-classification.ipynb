## Classification

Q. Find Median of crime per capita in the whole data

```r
mean(crim)
```

Q. Make a new column crime01: If a row has crim less than median, then crime01 = 0, otherwise, crime01 = 1

```r
Boston$crime01 <- 0
Boston$crime01[crim>mean(crim)] <- 1
Boston
```

Q. Separete train/set with ratio 2:1

```r
Boston <- na.omit(Boston)
```

```r
set.seed(1)
train_index <- sample(1:nrow(Boston), 0.7 * nrow(Boston))
test_index <- setdiff(1:nrow(Boston), train_index)
train = Boston[train_index,]
test = Boston[test_index,]
dim(train)
dim(test)
```

Q. Use lda (Linear Discriminant Analysis) to predict crime01 based on all data in Boston except crim

```r
?lda
```

```r
train
```

```r
class1 <- lda(crime01 ~ .-crim, data = train)
class1
```

```r
pred <- predict(class1, test)
names(pred)
```

```r
# attach(test)
```

```r
predmodel.test.lda = predict(class1, newdata=test)
table(Predicted=predmodel.test.lda$class, crime01=test$crime01)
```

```r
par(mfrow=c(1,1))
plot(predmodel.test.lda$x[,1], predmodel.test.lda$class, col=test$crime01+10)
```

Q. Try to remove some columns and use again lda to see whether you can improve the accuracy (you can get 11.9 %)

```r
class2 <- lda(crime01 ~ .-crim-chas-age, data = train)
class2
```

```r
pred <- predict(class2, test)
mean(pred$class == test$crime01)*100
```

Try qda (Quadratic Discriminant Analysis) to see whether it is better than lda

```r
df <- na.omit(obama_vs_mccain)
df <- subset(df, select = -c(State, Region))
df
```

Q. Predict Obama or McCain win (note: this dataset is quite small)

```r
df$Win <- ifelse(df$Obama > df$McCain, "Obama", ifelse(df$Obama < df$McCain, "McCain", "Draw"))   
```

```r
set.seed(1)
train_index <- sample(1:nrow(df), 0.7 * nrow(df))
test_index <- setdiff(1:nrow(df), train_index)
df_train = df[train_index,]
df_test = df[test_index,]
dim(df_train)
dim(df_test)
```

```r
lda_win = lda(Win~., data = df_train)
lda_win
```

```r
pred <- predict(lda_win, df_test)
mean(pred$class == df_test$Win)*100
```
